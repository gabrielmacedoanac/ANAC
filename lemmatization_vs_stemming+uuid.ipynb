#Criação do DataFrame de palavras

```python
import pandas as pd
```

```python
df_palavras = pd.DataFrame(['amigos', 'amigas', 'amizade', 'carreira', 'carreiras', 'tempestade'], columns=['Original'])
df_palavras
```

#Stemmer

```python
import nltk
nltk.download('rslp')
```

```python
stemmer = nltk.stem.RSLPStemmer()
```

```python
df_palavras['nltk_stemmer'] = [stemmer.stem(palavra) for palavra in df_palavras['Original']]
df_palavras
```

#Spacy lemmatization

```python
!python -m spacy download pt_core_news_sm
```

```python
import spacy
```

```python
nlp = spacy.load('pt_core_news_sm')
```

```python
doc = nlp(str([palavra for palavra in df_palavras['Original']]))
df_palavras['spacy_lemma'] = [token.lemma_ for token in doc if token.pos_ == 'NOUN']
df_palavras
```

```python
# Criar uuid
from uuid import uuid4
df_palavras['uuid'] = df_palavras.index.map(lambda _: uuid4())
df_palavras
```
